So i am about to create a simple neural network in Tensorflow Keras. In which i will give a set of Inputs xs and output ys. The neural network will automatically finds the pattern between the xs and ys.

```python
# Necessary Imports
import tensorflow as tf
import numpy as np
```

```python
#Declaring model inputs and outputs for training
xs= np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys= np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
#Building a simple Sequential model. It has 1 layer with 1 neuron
model= tf.keras.Sequential([

      #Define the input shape
      tf.keras.Input(shape=(1,)),

      #Add a dense layer
      tf.keras.layers.Dense(units=1)
])
```

```python
# Compile the model. For this you have to use two functions, the first one is the loss function and the second one is the optimizer function.
#here we use mean_squared_error as a lose function and  sgd (stochastic gradient descent) as optimizer

model.compile(loss='mean_squared_error', optimizer='sgd')
```

**Training the Neural** **Network**

```python
# Training is the Phase where where model learn the pattern between inputs and outputs in specified number of epoches

model.fit(xs,ys, epochs=500)
```

```python
#Now its time for prediction.

print(f'model Predicted : {model.predict(np.array([10.0]), verbose=0).item():5f}')
```
